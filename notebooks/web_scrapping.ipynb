```python
import os
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options

# Configurações do Selenium
def configure_driver(download_dir):
    """
    Configura o WebDriver do Selenium para o Chrome, definindo o diretório de download.

    :param download_dir: Diretório onde os arquivos baixados serão salvos.
    :return: Instância configurada do WebDriver.
    """
    chrome_options = Options()
    chrome_options.add_experimental_option('prefs', {
        "download.default_directory": download_dir,  # Define o diretório de download
        "download.prompt_for_download": False,  # Desabilita o prompt de download
        "download.directory_upgrade": True,
        "safebrowsing.enabled": True
    })
    service = Service(executable_path=r'C:\Users\breno\AppData\Local\Programs\Python\Python312/chromedriver.exe')  # Caminho para o executável do ChromeDriver
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

# Função para acessar o site do INMET e baixar os arquivos
def download_inmet_data(driver):
    """
    Acessa o site do INMET e baixa todos os arquivos de dados históricos disponíveis.

    :param driver: Instância do WebDriver configurada.
    """
    driver.get("https://portal.inmet.gov.br/dadoshistoricos")

    # Aguarda até que os elementos da tabela estejam visíveis
    WebDriverWait(driver, 20).until(
        EC.visibility_of_element_located((By.CLASS_NAME, "container"))
    )

    # Encontra todos os links de download na página
    download_links = driver.find_elements(By.XPATH, "//a[contains(@href, 'http://sinda.crn.inmet.gov.br')]")

    for link in download_links:
        # Clica em cada link para iniciar o download
        link.click()
        time.sleep(2)  # Aguarda 2 segundos entre os downloads para evitar sobrecarga

# Função principal
def main():
    """
    Função principal que configura o WebDriver, define o diretório de download e
    executa o download dos dados históricos do INMET.
    """
    download_dir = os.path.join(os.getcwd(), "inmet_data")  # Define o diretório de download
    os.makedirs(download_dir, exist_ok=True)  # Cria o diretório, se não existir

    # Configura o WebDriver
    driver = configure_driver(r'C:\Users\breno\OneDrive\Documentos\Faculdade\FIAP\Challenge\2024\nimbus\data\raw')
    
    try:
        # Realiza o download dos dados
        download_inmet_data(driver)
    finally:
        # Encerra o WebDriver
        driver.quit()

if __name__ == "__main__":
    main()
```
